```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd

from multioutreg.performance.metrics_generalized_api import (
    get_uq_performance_metrics_flexible,
)
from multioutreg.figures.performance_metric_figures import (
    plot_uq_metrics_bar,
    plot_predictions,
)

# Generate toy multioutput dataset
rng = np.random.RandomState(0)
X = rng.rand(300, 5)
y = np.dot(X, rng.rand(5, 3)) + rng.randn(300, 3) * 0.1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Fit a multioutput regressor
base_gp = GaussianProcessRegressor(random_state=0)
model = MultiOutputRegressor(base_gp)
model.fit(X_train, y_train)

# Gather per-output uncertainties
stds = []
for est in model.estimators_:
    _, s = est.predict(X_test, return_std=True)
    stds.append(s)
y_std = np.vstack(stds).T
# Avoid zeros in std array
y_std[y_std <= 0] = 1e-8

# Compute uncertainty metrics
try:
    metrics_df, overall = get_uq_performance_metrics_flexible(
        model,
        X_test,
        y_test,
        y_pred_std=y_std,
    )
    print("Overall metrics:", overall)
except AssertionError:
    # Fallback simple metrics if uncertainty-toolbox fails on this small toy data
    residuals = y_test - model.predict(X_test)
    mae = np.mean(np.abs(residuals), axis=0)
    rmse = np.sqrt(np.mean(residuals ** 2, axis=0))
    metrics_df = pd.DataFrame({"mae": mae, "rmse": rmse, "output": range(mae.size)})

# Plot per-output metrics and predictions
plot_uq_metrics_bar(metrics_df)
plt.show()

plot_predictions(model, X_test, y_test)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.inspection import partial_dependence, PartialDependenceDisplay
import math

# ----------- CONFIG -----------
n_targets = 4   # <-- Change this for any N
n_samples = 200
n_features = 1
random_state = 42

# ----------- DATA -----------
X, y = make_regression(
    n_samples=n_samples, n_features=n_features, n_targets=n_targets, noise=0.2, random_state=random_state
)

# Add interesting nonlinearity
for t in range(n_targets):
    if t % 3 == 0:
        y[:, t] = np.sin(X[:, 0] * (t+1)) + 0.1 * np.random.randn(n_samples)
    elif t % 3 == 1:
        y[:, t] = np.cos(X[:, 0] * (t+1)) + 0.1 * np.random.randn(n_samples)
    else:
        y[:, t] = np.tanh(X[:, 0] * (t+1)) + 0.1 * np.random.randn(n_samples)

# ----------- MODEL FIT -----------
kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))
gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=3, alpha=0.05)
multi_gpr = MultiOutputRegressor(gpr)
multi_gpr.fit(X, y)

```

```python
# ----------- PREDICT WITH UNCERTAINTY -----------
X_test = np.linspace(X.min(), X.max(), 200).reshape(-1, n_features)
means = []
stds = []
for estimator in multi_gpr.estimators_:
    mean, std = estimator.predict(X_test, return_std=True)
    means.append(mean)
    stds.append(std)
means = np.stack(means, axis=1)  # shape: (n_samples, n_outputs)
stds = np.stack(stds, axis=1)
```

```python
# ----------- PLOT PREDICTIONS WITH UNCERTAINTY -----------
ncols = 2
nrows = math.ceil(n_targets / ncols)
fig, axs = plt.subplots(nrows, ncols, figsize=(6 * ncols, 4 * nrows), squeeze=False)
output_names = [f"Output {i+1}" for i in range(n_targets)]

for i in range(n_targets):
    r, c = divmod(i, ncols)
    ax = axs[r][c]
    ax.scatter(X[:, 0], y[:, i], color='tab:blue', label='Train Data', alpha=0.5)
    ax.plot(X_test[:, 0], means[:, i], color='tab:red', label='GP Mean')
    ax.fill_between(
        X_test[:, 0],
        means[:, i] - 2 * stds[:, i],
        means[:, i] + 2 * stds[:, i],
        color='tab:red', alpha=0.2, label='2 Std Dev'
    )
    ax.set_title(output_names[i])
    ax.legend()
for j in range(i+1, nrows*ncols):
    axs[divmod(j, ncols)].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# ----------- PARTIAL DEPENDENCE PLOTS -----------
fig, axs = plt.subplots(nrows, ncols, figsize=(6 * ncols, 4 * nrows), squeeze=False)
for i in range(n_targets):
    r, c = divmod(i, ncols)
    PartialDependenceDisplay.from_estimator(
        multi_gpr, X, features=[0], target=i, ax=axs[r][c]
    )
    axs[r][c].set_title(f'PDP for {output_names[i]}')
for j in range(i+1, nrows*ncols):
    axs[divmod(j, ncols)].set_visible(False)
plt.tight_layout()
plt.show()
```
