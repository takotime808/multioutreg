## MultiOutput SHAP Plots

```python
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

from multioutreg.figures.shap_multioutput import plot_multioutput_shap_bar_subplots

n_targets = 3
n_features = 2
X, Y = make_regression(n_samples=200, n_features=n_features, n_targets=n_targets, noise=0.3, random_state=42)
regr = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42)).fit(X, Y)
output_names = [f"Target_{i}" for i in range(n_targets)]
feature_names = [f"F_{i}" for i in range(n_features)]

fig3 = plot_multioutput_shap_bar_subplots(regr, X, feature_names=feature_names, output_names=output_names)
```
